```python
!pip install diffusers transformers torch accelerate --quiet

from diffusers import StableDiffusionPipeline # the main pipeline class from HF
import torch
from IPython.display import display # helps render the images

# load pretrained diffusion model from HF
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16) # torch type to save memory
# move the pipeline to GPU for faster generation
pipe = pipe.to("cuda")
```

```python
prompt = "a town in the mountains at sunset"
image = pipe(prompt).images[0]
display(image)
```
